```python
# -----------------------------------------
# MENU (copied exactly as provided)
# -----------------------------------------
menu = {
    "Paneer Tikka":   {"category": "Starters",  "price": 180.0, "available": True},
    "Chicken Wings":  {"category": "Starters",  "price": 220.0, "available": False},
    "Veg Soup":       {"category": "Starters",  "price": 120.0, "available": True},
    "Butter Chicken": {"category": "Mains",     "price": 320.0, "available": True},
    "Dal Tadka":      {"category": "Mains",     "price": 180.0, "available": True},
    "Veg Biryani":    {"category": "Mains",     "price": 250.0, "available": True},
    "Garlic Naan":    {"category": "Mains",     "price":  40.0, "available": True},
    "Gulab Jamun":    {"category": "Desserts",  "price":  90.0, "available": True},
    "Rasgulla":       {"category": "Desserts",  "price":  80.0, "available": True},
    "Ice Cream":      {"category": "Desserts",  "price": 110.0, "available": False},
}

# -----------------------------------------
# CART LOGIC
# -----------------------------------------
cart = []

def print_cart():
    if not cart:
        print("Cart is empty.\n")
        return
    print("Current Cart:")
    for entry in cart:
        print(f"- {entry['item']} x{entry['quantity']}  (₹{entry['price']:.2f} each)")
    print()

def add_to_cart(item_name, quantity):
    if item_name not in menu:
        print(f"❌ '{item_name}' does not exist in the menu.\n")
        return

    if not menu[item_name]["available"]:
        print(f"❌ '{item_name}' is currently unavailable.\n")
        return

    # If already in cart → update quantity
    for entry in cart:
        if entry["item"] == item_name:
            entry["quantity"] += quantity
            print(f"Updated quantity of '{item_name}' to {entry['quantity']}.\n")
            return

    # Add new entry
    cart.append({
        "item": item_name,
        "quantity": quantity,
        "price": menu[item_name]["price"]
    })
    print(f"Added '{item_name}' x{quantity} to cart.\n")

def remove_from_cart(item_name):
    for entry in cart:
        if entry["item"] == item_name:
            cart.remove(entry)
            print(f"Removed '{item_name}' from cart.\n")
            return
    print(f"❌ '{item_name}' is not in the cart.\n")

# -----------------------------------------
# SIMULATION STEPS
# -----------------------------------------
print("STEP 1: Add Paneer Tikka × 2")
add_to_cart("Paneer Tikka", 2)
print_cart()

print("STEP 2: Add Gulab Jamun × 1")
add_to_cart("Gulab Jamun", 1)
print_cart()

print("STEP 3: Add Paneer Tikka × 1 (should update quantity)")
add_to_cart("Paneer Tikka", 1)
print_cart()

print("STEP 4: Try to add Mystery Burger (invalid)")
add_to_cart("Mystery Burger", 1)
print_cart()

print("STEP 5: Try to add Chicken Wings (unavailable)")
add_to_cart("Chicken Wings", 1)
print_cart()

print("STEP 6: Remove Gulab Jamun")
remove_from_cart("Gulab Jamun")
print_cart()

# -----------------------------------------
# FINAL ORDER SUMMARY
# -----------------------------------------
print("========== Order Summary ==========")

subtotal = 0
for entry in cart:
    line_total = entry["quantity"] * entry["price"]
    subtotal += line_total
    print(f"{entry['item']:<18} x{entry['quantity']}   ₹{line_total:.2f}")

print("------------------------------------")
gst = subtotal * 0.05
total = subtotal + gst

print(f"Subtotal:{'':>18}₹{subtotal:.2f}")
print(f"GST (5%):{'':>20}₹{gst:.2f}")
print(f"Total Payable:{'':>12}₹{total:.2f}")
print("====================================")














```

```python
import copy

# -----------------------------------------
# INVENTORY (exactly as provided)
# -----------------------------------------
inventory = {
    "Paneer Tikka":   {"stock": 10, "reorder_level": 3},
    "Chicken Wings":  {"stock":  8, "reorder_level": 2},
    "Veg Soup":       {"stock": 15, "reorder_level": 5},
    "Butter Chicken": {"stock": 12, "reorder_level": 4},
    "Dal Tadka":      {"stock": 20, "reorder_level": 5},
    "Veg Biryani":    {"stock":  6, "reorder_level": 3},
    "Garlic Naan":    {"stock": 30, "reorder_level": 10},
    "Gulab Jamun":    {"stock":  5, "reorder_level": 2},
    "Rasgulla":       {"stock":  4, "reorder_level": 3},
    "Ice Cream":      {"stock":  7, "reorder_level": 4},
}

# -----------------------------------------
# 1. Deep Copy
# -----------------------------------------
inventory_backup = copy.deepcopy(inventory)

print("Original Inventory:")
print(inventory)
print("\nBackup Inventory:")
print(inventory_backup)

# -----------------------------------------
# 2. Modify inventory to prove deep copy works
# -----------------------------------------
print("\nChanging inventory['Paneer Tikka']['stock'] = 5...\n")
inventory["Paneer Tikka"]["stock"] = 5

print("Inventory After Change:")
print(inventory)

print("\nBackup Inventory (Should be unchanged):")
print(inventory_backup)

# -----------------------------------------
# 3. Restore inventory
# -----------------------------------------
inventory = copy.deepcopy(inventory_backup)
print("\nInventory Restored:")
print(inventory)

# -----------------------------------------
# 4. Final cart from Task 2
# -----------------------------------------
cart = [
    {"item": "Paneer Tikka", "quantity": 3, "price": 180.0}
]

# -----------------------------------------
# 5. Deduct stock
# -----------------------------------------
print("\n=== Deducting Stock ===")

for entry in cart:
    item = entry["item"]
    qty = entry["quantity"]

    if item in inventory:
        available = inventory[item]["stock"]

        if available >= qty:
            inventory[item]["stock"] -= qty
            print(f"{item}: Deducted {qty}, remaining {inventory[item]['stock']}")
        else:
            print(f"⚠ Not enough stock for {item}. Only {available} available.")
            inventory[item]["stock"] = 0
    else:
        print(f"⚠ {item} not found in inventory.")

# -----------------------------------------
# 6. Reorder Alerts
# -----------------------------------------
print("\n=== Reorder Alerts ===")
for item, data in inventory.items():
    if data["stock"] <= data["reorder_level"]:
        print(f"⚠ Reorder Alert: {item} — Only {data['stock']} left (reorder level: {data['reorder_level']})")

# -----------------------------------------
# 7. Final Comparison
# -----------------------------------------
print("\nFinal Inventory:")
print(inventory)

print("\nBackup Inventory (Unchanged):")
print(inventory_backup)






























```

```python
# -----------------------------------------
# SALES LOG (exactly as given)
# -----------------------------------------
sales_log = {
    "2025-01-01": [
        {"order_id": 1, "items": ["Paneer Tikka", "Garlic Naan"], "total": 220.0},
        {"order_id": 2, "items": ["Gulab Jamun", "Veg Soup"], "total": 210.0},
        {"order_id": 3, "items": ["Butter Chicken", "Garlic Naan"], "total": 360.0},
    ],
    "2025-01-02": [
        {"order_id": 4, "items": ["Dal Tadka", "Garlic Naan"], "total": 220.0},
        {"order_id": 5, "items": ["Veg Biryani", "Gulab Jamun"], "total": 340.0},
    ],
    "2025-01-03": [
        {"order_id": 6, "items": ["Paneer Tikka", "Rasgulla"], "total": 260.0},
        {"order_id": 7, "items": ["Butter Chicken", "Veg Biryani"], "total": 570.0},
        {"order_id": 8, "items": ["Garlic Naan", "Gulab Jamun"], "total": 130.0},
    ],
    "2025-01-04": [
        {"order_id": 9, "items": ["Dal Tadka", "Garlic Naan", "Rasgulla"], "total": 300.0},
        {"order_id": 10, "items": ["Paneer Tikka", "Gulab Jamun"], "total": 270.0},
    ],
}

# -----------------------------------------
# 1. Print total revenue per day
# -----------------------------------------
print("=== Revenue Per Day ===")
revenue_per_day = {}

for date in sales_log:
    total = 0
    for order in sales_log[date]:
        total += order["total"]
    revenue_per_day[date] = total
    print(date, "→ ₹", total)

# -----------------------------------------
# 2. Best-selling day
# -----------------------------------------
best_day = None
best_revenue = 0

for date in revenue_per_day:
    if revenue_per_day[date] > best_revenue:
        best_revenue = revenue_per_day[date]
        best_day = date

print("\nBest-Selling Day:", best_day, "→ ₹", best_revenue)

# -----------------------------------------
# 3. Most ordered item
# -----------------------------------------
item_count = {}

for date in sales_log:
    for order in sales_log[date]:
        for item in order["items"]:
            if item not in item_count:
                item_count[item] = 1
            else:
                item_count[item] += 1

# Find item with highest count
most_item = None
most_count = 0

for item in item_count:
    if item_count[item] > most_count:
        most_item = item
        most_count = item_count[item]

print("\nMost Ordered Item:", most_item, "→", most_count, "orders")

# -----------------------------------------
# 4. Add new day
# -----------------------------------------
sales_log["2025-01-05"] = [
    {"order_id": 11, "items": ["Butter Chicken", "Gulab Jamun", "Garlic Naan"], "total": 490.0},
    {"order_id": 12, "items": ["Paneer Tikka", "Rasgulla"], "total": 260.0},
]

# Recalculate revenue
print("\n=== Updated Revenue Per Day ===")
revenue_per_day = {}

for date in sales_log:
    total = 0
    for order in sales_log[date]:
        total += order["total"]
    revenue_per_day[date] = total
    print(date, "→ ₹", total)

# Updated best-selling day
best_day = None
best_revenue = 0

for date in revenue_per_day:
    if revenue_per_day[date] > best_revenue:
        best_revenue = revenue_per_day[date]
        best_day = date

print("\nUpdated Best-Selling Day:", best_day, "→ ₹", best_revenue)

# -----------------------------------------
# 5. Numbered list of all orders
# -----------------------------------------
print("\n=== All Orders (Numbered) ===")

count = 1
for date in sales_log:
    for order in sales_log[date]:
        items_text = ", ".join(order["items"])
        print(count, ". [", date, "] Order #", order["order_id"],
              " — ₹", order["total"], " — Items:", items_text, sep="")
        count += 1
```
